```python
# Function to find outliers
def find_outliers_IQR(data):
  q1 = data.quantile(0.25)
  q3 = data.quantile(0.75)
  IQR = q3 - q1
  outliers = data[((data<(q1-1.5*IQR))| (data>(q3+1.5*IQR)))]
  return outliers
```

```python
# Method to control the rows and columns truncation in dataset
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 150)
```

```python
# Call upon find_outliers_IQR function to detect outliers in Survival_Months
sm_outliers = find_outliers_IQR(data['Survival_Months'])
print("number of outliers: " + str(len(sm_outliers)))
sm_outliers
```

```python
# Remove outliers from Survival_Months variable
data.drop(index=sm_outliers.index, inplace=True)
```

```python
# Fill missing variable in Tumor_Size with mean value
Mean_Tumor_Size = data['Tumor_Size'].mean()
data['Tumor_Size'].fillna(Mean_Tumor_Size, inplace=True)
```

```python
# Fill missing variable in Regional_Node_Examined with mean value
Mean_Regional_Node_Examined = data['Regional_Node_Examined'].mean()
data['Regional_Node_Examined'].fillna(Mean_Regional_Node_Examined, inplace=True)
```

```python
# Perform case-wise deletion
data = data.dropna()
```

```python
# Codify Mortality_Status variable to binary categorical values
data['Mortality_Status'] = data['Mortality_Status'].map({'Alive':1, 'Dead':0})
```

```python
# Codify Sex variable to binary categorical values
data['Sex'] = data['Sex'].map({'Female':1, 'Male':0})
```

```python
# Codify T_Stage variable to binary categorical values
data['T_Stage'] = data['T_Stage'].map({'T1': 1, 'T2': 2, 'T3': 3, 'T4': 4})
```

```python
# Codify N_Stage variable to binary categorical values
data['N_Stage'] = data['N_Stage'].map({'N1': 1, 'N2': 2, 'N3': 3})
```

```python
# Codify 6th_Stage variable to binary categorical values
data['6th_Stage'] = data['6th_Stage'].map({'IIA': 1, 'IIB': 2, 'IIIA': 3, 'IIIB': 4, 'IIIC': 5})
```

```python
# Codify A_Stage variable to binary categorical values
data['A_Stage'] = data['A_Stage'].map({'Regional': 1, 'Distant': 2})
```

```python
# Codify Estrogen_Status variable to binary categorical values
data['Estrogen_Status'] = data['Estrogen_Status'].map({'Positive': 1, 'Negative': 0})
```

```python
# Codify Progesterone_Status variable to binary categorical values
data['Progesterone_Status'] = data['Progesterone_Status'].map({'Positive': 1, 'Negative': 0})
```

```python
# Convert 2 Float data types into integers
data = data.astype({"Age": "int", "Regional_Node_Examined": "int"})
```

```python
# Import preprocessing library from sklearn
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```
